## Aufgabe 2 (2 Punkte)
Lesen Sie die CSV-Datei bundeslaender.txt ein.

a) Erzeugen Sie eine neue Datei mit den Spaltennamen „land“, “area“, „female“, „male“, „ population“ und
„density“. Die density ist die Anzahl der Einwohner pro Quadratkilometer. (1 Punkt)

<img src="images/2a.png" width="1200">

```python
bundeslaender = pd.read_csv('bundeslaender.txt', sep=' ')

bundeslaender['population'] = bundeslaender['male'] + bundeslaender['female']
bundeslaender['density'] = bundeslaender['population'] / bundeslaender['area'] * 1000
# bundeslaender['density'] = bundeslaender['density'].round(0)
# bundeslaender['density'] = bundeslaender['density'].astype(int)

print(bundeslaender.head(16))
# bundeslaender.head(16)
```

b) Geben Sie alle Zeilen aus mit einer größeren weiblichen Bevölkerungszahl als die der Männer. Wie viele
sind es? 0,5 Punkte

```python
tmp = bundeslaender[(bundeslaender['female'] > bundeslaender['male'])]

print(tmp)
print(f"Es sind {tmp.shape[0]} Bundesländer mit mehr frauen als Männer")
```

c) Geben Sie alle Zeilen aus, deren Bevölkerungsdichte größer als 1000 Einwohner pro Quadratkilometer ist.
0,5 Punkte

<img src="images/2c.png" width="1000">

```python
tmp = bundeslaender[(bundeslaender['density'] > 1000)]

print(tmp)
print(f"Es sind {tmp.shape[0]} Bundesländer mit mehr als 1000 Einwohnern pro Quadratkilometer")
```

## Aufgabe 3 ( 1 Punkt)
a) Schreiben Sie ein Script und finden Sie heraus, wieviele Arbeitstage/reguläre Wochentage es zwischen dem 24. 12. 2025 und dem
6.01.2026 gibt – generieren Sie die unten angebene Ausgabe. (0,5 Punkte)

<img src="images/3a.png" width="1200">

```python
from datetime import date

start = date(2025, 12, 24)
end = date(2026, 1, 6)

weekdays = pd.date_range(start, end)
weekdays = weekdays[weekdays.weekday <= 4]
print(weekdays)
print(f"Es gibt {weekdays.shape[0]} normale Wochentage/reguläre Arbeitstage in den Weihnachtsferien 2025/26.")
```
