## Aufgabe 1 (1 Punkt)
Laden Sie den iris-Datensatz und zeichnen Sie vier Histogramme jeweils der vier numerischen Merkmale für die
drei Objektklassen setosa, versicolor, und virginica. Färben Sie nach den Objektklassen ein. Erzeugen Sie die
folgenden Ausgaben:

<img src="images/1.png" width="1200">

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("iris.csv")

# Filtering after classes
setosa = df[df["variety"] == "Setosa"]
versicolor = df[df["variety"] == "Versicolor"]
virginica = df[df["variety"] == "Virginica"]

# size of each diagram/figure (w,h)
plt.figure(figsize=(16,10))

# Histogram of sepal length
plt.subplot(3,3,1)
plt.hist(setosa["sepal.length"], label="setosa", color="blue")
plt.hist(versicolor["sepal.length"], label="versicolor", color="red")
plt.hist(virginica["sepal.length"], label="virginica", color="green")
plt.xlabel("sepal length (cm)")
plt.legend(loc="upper right")

# Histogram of petal width
plt.subplot(3,3,2)
plt.hist(setosa["petal.width"], label="setosa", color="blue")
plt.hist(versicolor["petal.width"], label="versicolor", color="red")
plt.hist(virginica["petal.width"], label="virginica", color="green")
plt.xlabel("petal.width (cm)")
plt.legend(loc="upper right")

# Histogram of petal length
plt.subplot(3,3,3)
plt.hist(setosa["petal.length"], label="setosa", color="blue")
plt.hist(versicolor["petal.length"], label="versicolor", color="red")
plt.hist(virginica["petal.length"], label="virginica", color="green")
plt.xlabel("petal length (cm)")
plt.legend(loc="upper right")

# Histogram of sepal width
plt.subplot(3,3,4)
plt.hist(setosa["sepal.width"], label="setosa", color="blue")
plt.hist(versicolor["sepal.width"], label="versicolor", color="red")
plt.hist(virginica["sepal.width"], label="virginica", color="green")
plt.xlabel("sepal.width (cm)")
plt.legend(loc="upper right")

# For correct alignment of the subplots (no overlap)
plt.tight_layout()
plt.show()
```

## Aufgabe 2 (1 Punkte)
a) Zeichnen Sie ein Streudiagramm für die Attribute Kelchblaltlänge in der x-Achse und Kelchblattbreite in der y-Achse, wie unten in der
Abblidung sowie für Blütenblattlänge und Blütenblattbreite auch wie unten in der Abbildung. (0.5 Punkt)

<img src="images/2a.png" width="1200">

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("iris.csv")
plt.figure(figsize=(12,6))

a = plt.subplot(1,2,1)
df.plot.scatter(x="sepal.length", y="sepal.width", marker="+", ax=a)
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
# x-axis starts before 4
a.set_xlim(3.5, 8.5)
# visible x-ticks
a.set_xticks([4, 5, 6, 7, 8])
a.set_ylim(1.5,5)
a.set_yticks([2.0, 2.5, 3.0, 3.5, 4.0, 4.5])

b = plt.subplot(1,2,2)
df.plot.scatter(x="petal.length", y="petal.width", color="green", ax=b)
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")

plt.tight_layout()

plt.show()
```

b) Berechnen Sie den Korrelationskoeffizient für die beiden Merkmalspaare aus Aufgabe 1 (Kelchblattlänge u. Kelchblattbreite,
Blütenblattlänge und Blütenblattbreite. Erzeugen Sie die Ausgaben (0.5 Punkt) :
- The correlation coefficient between petal length (cm) and petal width (cm) attributes is 0.96287.
- The correlation coefficient between petal length (cm) and petal width (cm) attributes is -0.11757.

```python
# measures how well points lie on a rising or falling straight line in a scatter
# sepal (Kelchblatt) and pedal (Blütenblatt)
print("Korrelationskoeffizient Kelchblattlänge und Kelchblattbreite: " + str(df['sepal.length'].corr(df['sepal.width'])))
print("Korrelationskoeffizient Blütenblattlänge und Blütenblattbreite: " + str(df['petal.length'].corr(df['petal.width'])))
```
