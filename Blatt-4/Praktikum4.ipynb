## Aufgabe 1 (1 Punkt)
Laden Sie den iris-Datensatz und zeichnen Sie vier Histogramme jeweils der vier numerischen Merkmale für die
drei Objektklassen setosa, versicolor, und virginica. Färben Sie nach den Objektklassen ein. Erzeugen Sie die
folgenden Ausgaben:

<img src="images/1.png" width="1200">

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('iris.csv')

def histogram(frame: pd.DataFrame, xlabel: str, value_col: str, category: str, ax) -> None:
    for label, group in frame.groupby([category]):
        ax.hist(group[value_col], label=label)
    ax.set_xlabel(xlabel)
    ax.legend()

fig, ax = plt.subplots(ncols=2, nrows=2)

histogram(df, 'sepal length (cm)', 'sepal.length', 'variety', ax[0][0])
histogram(df, 'sepal width (cm)', 'sepal.width', 'variety', ax[1][0])
histogram(df, 'petal length (cm)', 'petal.length', 'variety', ax[0][1])
histogram(df, 'petal width (cm)', 'petal.width', 'variety', ax[1][1])

plt.tight_layout()
plt.show()
```
